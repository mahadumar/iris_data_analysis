# Task 1: Exploring and Visualizing a Simple Dataset

## Problem Statement:

Understanding a dataset is the first and most important step in any data science or machine learning project. Before building predictive models, it's essential to explore the data, identify patterns, and visualize distributions to gain meaningful insights. In this task, we will work with the famous Iris dataset, which includes measurements of different flower species. The goal is to inspect the dataset structure and use basic data visualization techniques to analyze the relationships and trends in the data.

## Goal

The goal of this task is to:

- Learn how to load and inspect a dataset using `pandas`.
- Explore summary statistics to understand data characteristics.
- Visualize feature relationships, distributions, and potential outliers using `matplotlib` and `seaborn`.

This will help build a strong foundation in exploratory data analysis (EDA), a crucial skill in any data science workflow.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load the built-in iris dataset
iris = sns.load_dataset('iris')

# basic dataset overview
print("shape of the dataset:", iris.shape)
print("\ncolumn names:", iris.columns.tolist())

print("\nfirst five rows of the dataset:")
print(iris.head())

print("\ndataset info:")
iris.info()

print("\nstatistical summary:")
print(iris.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#set seaborn style for better visuals
sns.set(style="whitegrid")

#scatter plot to see relationship between sepal length and width
plt.figure(figsize=(7, 5))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", palette="Set2")
plt.title("sepal length vs sepal width")
plt.show()

#histograms for all numeric features
iris.hist(edgecolor='black', linewidth=1.2, figsize=(10, 6))
plt.suptitle("histograms of iris features", fontsize=16)
plt.tight_layout()
plt.show()

#box plots for each feature grouped by species
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

for feature in features:
    plt.figure(figsize=(7, 4))
    sns.boxplot(x="species", y=feature, data=iris, palette="Pastel1")
    plt.title(f"box plot of {feature} by species")
    plt.show()
```
